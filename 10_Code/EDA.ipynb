```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the custom color palette
custom_palette = sns.color_palette("dark:#79B473", as_cmap=True)

# Generate a list of colors from the colormap
colors = [custom_palette(i / 50) for i in range(50)]

fig, ax = plt.subplots(2, 1, figsize=(24, 12))

# Plot the bar plot for teams that have played the most number of seasons
sns.barplot(
    x=MTeams.sort_values(by="SeasonPlayed", ascending=False).head(50)["TeamName"],
    y=MTeams.sort_values(by="SeasonPlayed", ascending=False).head(50)["SeasonPlayed"],
    ax=ax[0],
    palette=colors,
)
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45)
ax[0].set_xlabel("Team Name", fontsize=12)
ax[0].set_ylabel("Number of Seasons Played", fontsize=12)
for container in ax[0].containers:
    ax[0].bar_label(
        container,
        padding=3,
        fontsize=12,
        color="black",
        label_type="edge",
        weight="bold",
    )
ax[0].set_title("Teams that have played the most number of seasons", fontsize=16)

# Plot the bar plot for teams that have played the least number of seasons
sns.barplot(
    x=MTeams.sort_values(by="SeasonPlayed", ascending=True).head(50)["TeamName"],
    y=MTeams.sort_values(by="SeasonPlayed", ascending=True).head(50)["SeasonPlayed"],
    ax=ax[1],
    palette=colors,
)
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=45)
ax[1].set_xlabel("Team Name", fontsize=12)
ax[1].set_ylabel("Number of Seasons Played", fontsize=12)
for container in ax[1].containers:
    ax[1].bar_label(
        container,
        padding=3,
        fontsize=12,
        color="black",
        label_type="edge",
        weight="bold",
    )
ax[1].set_title("Teams that have played the least number of seasons", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot 1: Mean Winning and Losing Scores over Seasons
fig, ax = plt.subplots(figsize=(12, 6))
avg_score = MNCAATourneyCompactResults.groupby("Season")[["WScore", "LScore"]].mean()
sns.lineplot(data=avg_score, ax=ax, palette={"WScore": "green", "LScore": "red"})
ax.axhline(
    y=avg_score["WScore"].mean(), color="darkgreen", linestyle="--", label="Mean WScore"
)
ax.axhline(
    y=avg_score["LScore"].mean(), color="darkred", linestyle="--", label="Mean LScore"
)
ax.set_title("Mean winning and losing scores over seasons", fontsize=16)
ax.set_xlabel("Season")
ax.set_ylabel("Score")
ax.legend()
ax.grid(True)
ax.text(
    0.5,
    -0.13,
    "Figure 1: Illustrates the trend of mean winning (WScore) and losing (LScore) scores across multiple seasons ",
    ha="center",
    fontsize=12,
    transform=ax.transAxes,
)

plt.show()
```

```python
# Plot 2: Difference between Mean Winning and Losing Scores over Seasons
fig, ax = plt.subplots(figsize=(12, 6))
avg_score["ScoreDifference"] = avg_score["WScore"] - avg_score["LScore"]
sns.lineplot(data=avg_score["ScoreDifference"], ax=ax, color="gray")
ax.axhline(
    y=avg_score["ScoreDifference"].mean(),
    color="black",
    linestyle="--",
    label="Mean Difference",
)
ax.set_title(
    "Difference between mean winning and losing scores over seasons", fontsize=16
)
ax.set_xlabel("Season")
ax.set_ylabel("Score Difference")
ax.legend()
ax.grid(True)
ax.text(
    0.5,
    -0.13,
    "Figure 2: Illustrates the fluctuation in the difference between the mean winning (WScore) and losing (LScore) scores across multiple NCAA tournament seasons",
    ha="center",
    fontsize=12,
    transform=ax.transAxes,
)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams.update({"font.size": 14})

fig, ax = plt.subplots(1, 2, figsize=(24, 6))

# For MSecondaryTourneyCompactResults
counts = (
    MRegularSeasonCompactResults["WLoc"]
    .value_counts()
    .rename(index={"H": "Home", "A": "Away", "N": "Neutral"})
)
percentages = (
    MRegularSeasonCompactResults["WLoc"]
    .value_counts(normalize=True)
    .rename(index={"H": "Home", "A": "Away", "N": "Neutral"})
    * 100
)

sns.barplot(x=counts.index, y=counts.values, ax=ax[0], palette="Blues_d")
ax[0].set_title("Counts of location of the winning team (men's teams)")
ax[0].set_xlabel("Location of the winning team")
ax[0].set_ylabel("Count")
for i, v in enumerate(counts.values):
    ax[0].text(i, v + 50, f"{percentages.values[i]:.2f}%", color="black", ha="center")
fig.text(
    0.20,
    -0.04,
    "Figure 3: Distribution of winning locations for men's teams ",
    ha="center",
    va="center",
    fontsize=15,
)

# For WRegularSeasonCompactResults
counts = (
    WRegularSeasonCompactResults["WLoc"]
    .value_counts()
    .rename(index={"H": "Home", "A": "Away", "N": "Neutral"})
)
percentages = (
    WRegularSeasonCompactResults["WLoc"]
    .value_counts(normalize=True)
    .rename(index={"H": "Home", "A": "Away", "N": "Neutral"})
    * 100
)

sns.barplot(x=counts.index, y=counts.values, ax=ax[1], palette="Reds_d")
ax[1].set_title("Counts of location of the winning team (women's teams)")
ax[1].set_xlabel("Location of the winning team")
ax[1].set_ylabel("Count")
for i, v in enumerate(counts.values):
    ax[1].text(i, v + 50, f"{percentages.values[i]:.2f}%", color="black", ha="center")
fig.text(
    0.70,
    -0.04,
    "Figure 4: Distribution of winning locations for women's teams ",
    ha="center",
    va="center",
    fontsize=15,
)

plt.tight_layout()
plt.show()
```
